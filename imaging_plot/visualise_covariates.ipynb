### jump frequency model evidence as covariate

```python
import numpy as np
from nilearn import plotting
from IPython.display import Image
from nilearn.glm import threshold_stats_img
```

```python
bg = '../../../TSL_anat/mean_anat.nii.gz'
```

```python
def plot_overlay(map1, map2, display_mode='z', peaks=[8, 18, 33, 43, 65]):
    """plotting overlaid of 2 contrasts"""
    cmp_l = plotting.cm.black_green_r # green - map1
    cmp_h = plotting.cm.black_pink_r # pink - map2
    
    map_1, threshold = threshold_stats_img(
    map1, alpha=.001, height_control='fpr', cluster_threshold=10)
    display4 = plotting.plot_stat_map(map_1, threshold=threshold, vmax=7,
                  display_mode=display_mode, cut_coords=(peaks), draw_cross=False, bg_img=bg,
                                    cmap=cmp_l)
    # add overlay
    try:
        map_2, threshold = threshold_stats_img(
            map2, alpha=.001, height_control='fpr', cluster_threshold=10)
        display4.add_overlay(map_2, threshold=threshold, cmap=cmp_h)
    except Exception:
        pass
    # save
    display4.savefig('./figs/hl_lh_covar_overlay_z.png')
```

Jump frequency model posterior mean

```python
model_dir = '../../../TSL_output/2ndLevel_io_jf_hl_pp_pe_sd_covariate_FDR0001'
img1 = f'{model_dir}/2ndLevel/con_0001/spmT_0001_thr.nii' # posterior mean of low
img2 = f'{model_dir}/2ndLevel/con_0002/spmT_0001_thr.nii' # high
```

```python
plot_overlay(img1, img2, display_mode='z', peaks=[8, 18, 33, 43, 65])
```

```python
plot_overlay(img1, img2, display_mode='y', peaks=[-14, -7, 0, 18])
```

```python
plot_overlay(img1, img2, display_mode='x', peaks=[-62, 0, 28, 40])
```
